```python
import pandas as pd
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/IRIS.csv')
df.head()
```

```python
# prompt: Ploting bar graph for the data set

plt.figure(figsize=(12,8))
plt.title('Bar Graph')
sns.countplot(x='species', data=df)
plt.show()
```

```python
# prompt: Plotting line chart

plt.figure(figsize=(12,6))
plt.title('Line Graph')
sns.lineplot(x='sepal_length', y='sepal_width', data=df)
plt.show()
```

```python
# prompt: Box plot

plt.figure(figsize=(10,5))
plt.title('Box Plot')
sns.boxplot(x='species', y='sepal_length', data=df)
plt.show()
```

```python
# prompt: Pair Plot Graph

plt.figure(figsize=(10,5))
plt.title('Pair Plot')
sns.pairplot(df, hue='species',)
plt.show()
```

```python
# prompt: Shwo groups with different colours using hue and tip size with size


plt.figure(figsize=(12,8))
plt.title('Bar Graph')
sns.countplot(x='species', data=df)
plt.show()

plt.figure(figsize=(12,6))
plt.title('Line Graph')
sns.lineplot(x='sepal_length', y='sepal_width', data=df)
plt.show()

plt.figure(figsize=(10,5))
plt.title('Box Plot')
sns.boxplot(x='species', y='sepal_length', data=df)
plt.show()

plt.figure(figsize=(10,5))
plt.title('Pair Plot')
sns.pairplot(df, hue='species',)
plt.show()

# Scatter plot with hue and size
plt.figure(figsize=(10,6))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', size='petal_width', data=df)
plt.title('Scatter Plot with Hue and Size')
plt.show()
```

```python
# prompt: violin graph

plt.figure(figsize=(10,5))
plt.title('Violin Plot')
sns.violinplot(x='species', y='sepal_length', data=df)
plt.show()
```
